```python
from itertools import product #cartesian product
from math import sqrt, cos, sin, atan2
import numpy as np

def z2y(r, x): # converts impedance Z=R+jX to admittance Y=G+jB
    return r/(r**2+x**2), -x/(r**2+x**2)

def p1t(g,b,v1,v2,p2):
    try:
        temp = sqrt(b**2*(-p2**2+(2*g*p2+(b**2+g**2)*v1**2)*v2**2-g**2*v2**4))
    except ValueError:
        return nan, nan
    t = -arccos((-g*p2+g**2*v2**2+temp)/((b**2+g**2)*v1*v2))
    p1 = 1/(b**2+g**2) * ((-b**2+g**2)*p2+g**3*(v1**2-v2**2)+b**2*g*(v1**2+v2**2)-2*g*temp)
    return p1, t
def e2f2_(b, g, p2, e1, p1):
    try:
        temp = sqrt(g*(-g*p1**2+(b**2+g**2)*e1**2*(p1+p2)))
    except ValueError:
        return np.nan, np.nan
    A = g*(b**2+g**2)*e1**2-g**2*p1
    B = b*temp
    C = g*(b**2+g**2)*e1
    e2 = (A+B)/C 
    #e2 = (A-B)/C
    A = b*p1
    B = temp
    C = (b**2+g**2)*e1
    f2 = (A+B)/C 
    #f2 = (A-B)/C
    return e2, f2
def v2_(b, g, p2, e1, p1):
    e2, f2 = e2f2_(b, g, p2, e1, p1)
    return sqrt(e2**2+f2**2)
def v2t_(b, g, p2, e1, p1):
    e2, f2 = e2f2_(b, g, p2, e1, p1)
    return sqrt(e2**2+f2**2), arctan2(f2, e2)
```

```python
candidates = p0v0v1p2v2v3v4
```

```python
def check(b,g,p1,p2,v1,v2,t,tol=1e-4):
    t *= pi/180
    p1_hat = g*(v1**2-v1*v2*cos(t)) + b*v1*v2*sin(t)
    p2_hat = g*(v2**2-v1*v2*cos(t)) - b*v1*v2*sin(t)
    err1 = abs((p1_hat-p1)/p1_hat)
    err2 = abs((p2_hat-p2)/p2_hat)
    return max([err1, err2])
```

#The recipe

1. Start at the parents of the deepest leaves. The leaves have only a fixed demand, not a fixed voltage. So enumerate $(P_i, V_i)$ for the leaf parents, but only keep those which produce a feasible flow, given the demands at the leaves.
2. In general as we recurse, the voltage magnitude at the 'leaves' will also be fixed. In this case, we only enumerate the $V_i$ at the leaf parents, keeping those which produce a real $P_i$ while satisfying the demands.

**<span style='color:red'>Actually we need to know the amount of power the leaf parents send to each leaf individually.</span>** So we enumerate $P_i$ at each leaf, then enumerate $V_i$ at the leaf parents.

###Another way of looking at it

    | G0--L1
    |  \
    |  L2
    
If only the demands at L1 and L2 are fixed, then all of $(P_0, V_0, V_1, V_2, \theta_1, \theta_2)$ are free. Setting $(P_0, V_0)$ implies the others.

If both the demands and voltage magnitudes at L1 and L2 are fixed, then only $(P_0, V_0, \theta_1, \theta_2)$ are free. It's easiest to set $V_0$ because this immediately fixes the $\theta$s and implies how $P_0$ is split among its branches.
