The equations describing real power flow between two buses $k$ and $m$ are:

$$ P_k = g(V_k^2-V_k V_m \cos \theta_{km}) - b V_k V_m \sin \theta_{km} $$
$$ P_m = g(V_m^2-V_k V_m \cos \theta_{km}) + b V_k V_m \sin \theta_{km} $$

If $(P_m,V_m,V_k)$ are known, we can solve for $(P_k, \theta_{km})$:

$$P_k = \frac{\left(-b^2+g^2\right) P_m+g^3 \left(V_k^2-V_m^2\right)+b^2 g \left(V_k^2+V_m^2\right)-2 g \sqrt{D}}{b^2+g^2}$$

$$\theta_{km} = -\cos^{-1}\left[\frac{-g P_m+g^2 V_m^2 \pm \sqrt{D}}{\left(b^2+g^2\right) V_k V_m}\right]$$

$$D = b^2 \left(-P_m^2+\left(2 g P_m+\left(b^2+g^2\right) V_k^2\right) V_m^2-g^2 V_m^4\right) \geq 0$$

The solution of course is only valid when the discriminant $D$ is non-negative. Observe that there are two possible values for $\theta$.

##What the code is doing for now

I always use the '+' solution for $\theta$ which I'll explain in the next section. The power flow code I'm using right now looks like this:

```julia
def pk_t(g,b,vk,vm,pm):
    ''' given vk,vm,pm compute pk and theta (t) '''
    try:
        temp = sqrt(b**2*(-pm**2+(2*g*pm+(b**2+g**2)*vk**2)*vm**2-g**2*vm**4))
    except ValueError:
        return nan, nan
    t = -arccos((-g*pm+g**2*vm**2+temp)/((b**2+g**2)*vk*vm))
    pk = 1/(b**2+g**2) * ((-b**2+g**2)*pm+g**3*(vk**2-vm**2)+b**2*g*(vk**2+vm**2)-2*g*temp)
    return pk, t
```

If the discriminant is negative, it tries to take the square root which gives an error which is caught, and NaN (not a number) is returned. At another step I filter out the NaN solutions; if they accidentally slip through the filter they are pretty easy to spot since they screw up any further analysis.

In earlier versions of the code I used a cython version of this which is much faster, but currently it isn't the bottleneck.

##Which value of $\theta$ to use

Recall there are two possible solutions for $\theta$:

$$\theta_{km} = -\cos^{-1}\left[\frac{-g P_m+g^2 V_m^2 \pm \sqrt{D}}{\left(b^2+g^2\right) V_k V_m}\right]$$

It turns out that the '+' solution always produces a smaller (in magnitude) angle and a more sensible solution. Why?

To get small angle differences, we want the argument to $\cos^{-1}$ to be close to 1. So let's examine that quantity.

```julia
x = linspace(-1,1,1000)
Series(arccos(x), index=x).plot(figsize=(6,4))
xlabel(r'$x$'); ylabel(r'$\cos^{-1}x$');
```

$$f = \frac{-g P_m+g^2 V_m^2\pm\sqrt{D}}{\left(b^2+g^2\right) V_k V_m}$$

Observe that the '+' solution will always be greater. So if we can show that $-1\leq f \leq 1$ then we've established that the '+' solution always gives a smaller angle.

```julia
def f(g=1,b=-10,vk=1,vm=1,pm=-1):
    temp = b**2*(-pm**2+(2*g*pm+(b**2+g**2)*vk**2)*vm**2-g**2*vm**4)
    if temp < 0:
        return None
    x1 = (-g*pm+g**2*vm**2+temp**.5)/((b**2+g**2)*vk*vm)
    x2 = (-g*pm+g**2*vm**2-temp**.5)/((b**2+g**2)*vk*vm)
    return x1, x2
```

```julia
def rand_solution():
    ''' generate a random set of arguments for f '''
    def r(a,b):
        return (b-a)*rand() + a
    #return dict(g=r(0,20),
    #         b=r(-50,50),
    #         vk=r(.5,1.5),
    #         vm=r(.5,1.5),
    #         pm=r(-10,0))
    return r(0,20), r(-50,50), r(.5,1.5), r(.5,1.5), r(-10,0)
```

Generate a bunch of random arguments for $f$ and look at the top and bottom values for both the + and - solutions.

```julia
%%time
n = 150000
plus_list, minus_list = [], []
for i in range(n):
    #solution = f(**rand_solution())
    solution = f(*rand_solution())
    if solution is not None:
        plus_list.append(solution[0])
        minus_list.append(solution[1])
print(len(solution_list)) # the rest were infeasible
print('Maximum + solution:', max(plus_list))
print('Minimum - solution:', max(minus_list))
```

So at least empirically we're observing what we want, that $-1\leq f \leq 1$

##An observation: $\theta$ always negative

I'm only looking at the angle difference between a 'generator' and a load:

$$\theta_{km} = \theta_k - \theta_m$$

where bus $k$ is injecting power into the line ($P_k>0$) and bus $m$ is using power ($P_m<0$). In this case it looks like we have that $\theta_{km} < 0$. Must that be the case? Why?

From looking at a few matpower cases it looks like this generally holds, but it'd be nice to understand what's going on.

```julia
def pk_t(g=1,b=-10,vk=1,vm=1,pm=-1):
    ''' given vk,vm,pm compute pk and theta (t) '''
    temp = b**2*(-pm**2+(2*g*pm+(b**2+g**2)*vk**2)*vm**2-g**2*vm**4)
    #if temp < 0:
    #    return nan, nan
    temp **= .5
    t = -arccos((-g*pm+g**2*vm**2+temp)/((b**2+g**2)*vk*vm))
    pk = 1/(b**2+g**2) * ((-b**2+g**2)*pm+g**3*(vk**2-vm**2)+b**2*g*(vk**2+vm**2)+2*g*temp)
#    pk = 1/(b**2+g**2) * ((-b**2+g**2)*pm+g**3*(vk**2-vm**2)+b**2*g*(vk**2+vm**2)-2*g*temp)
    return pk, t
def check(g=1,b=-10,pk=1.01,vk=1,vm=1,pm=-1,t=-.1):
    ''' sanity check after a solution has been found '''
    pk_hat = g*(vk**2-vk*vm*cos(t)) + b*vk*vm*sin(t)
    pm_hat = g*(vm**2-vk*vm*cos(t)) - b*vk*vm*sin(t)
    err1 = abs((pk_hat-pk)/pk_hat)
    err2 = abs((pm_hat-pm)/pm_hat)
    #return array([vstack([x,y]).max(axis=0) for x,y in zip(err1,err2)])
    #return vstack([err1,err2]).max(axis=0)
    return max([err1, err2])
def disc(g=1,b=-10,vk=1,vm=1,pm=-1):
    return b**2*(-pm**2+(2*g*pm+(b**2+g**2)*vk**2)*vm**2-g**2*vm**4)
```
